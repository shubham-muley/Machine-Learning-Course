## Simple Model

```python
import sklearn.datasets
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
cancer_ds=sklearn.datasets.load_breast_cancer()
X=cancer_ds.data                     #Separating data and target also Converting into ndarray
Y=cancer_ds.target
print(X.shape, Y.shape)
```

```python
data = pd.DataFrame(cancer_ds.data, columns=cancer_ds.feature_names)
data['Class'] = cancer_ds.target         #appending new column "Class" to the data with its values
data.head()
```

```python
data.describe()
```

```python
print(data['Class'].value_counts()) #Counting the Binary values
```

```python
print(cancer_ds.target_names)
```

```python
data.groupby('Class').mean()
```

```python
X = data.drop('Class', axis=1)             #without column class
Y= data['Class']
```

## random_state attribute of train_test_split function
The same data is gone under train and test we can see that in the below demonstration taking only 0.2% data as test data so that we can easily compare between the data

```python
print("The Output format is\nIndex no.  Label")
for _ in range(4):
    X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.002, random_state=1)#0.2% Testdata
    print(Y_test)
```

## stratify attribute of train_test_split function
The distribution of the data between training and testing is same such that its mean, SD etc are same

```python
for i in range(5):
    X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2,stratify = Y)
    print(Y_test.mean())

print("*"*30)

for i in range(5): 
    X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2)
    print(Y_test.mean())
```

```python
#Binarization of Input

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.25,stratify = Y)    #25% test data  
                                                                                #uniform distribution of train and test data

X_binarised_train = X_train.apply(pd.cut, bins=2, labels = [1,0])
X_binarised_test = X_test.apply(pd.cut, bins=2, labels = [1,0])
type(X_binarised_test)

X_binarised_train=X_binarised_train.values
X_binarised_test=X_binarised_test.values

type(X_binarised_test)
```

# Prediction model using loop

## On Train data

```python
b=0
accurate_rows = 0
max_accurate_rows=0
b_optimum=0
print("For Train data:\n")
print("OUTPUT FORMAT iS \n{b ,Accurate_Rows ,Accuracy }")
for b in range(1,31):
    y_pred=0
    accurate_rows=0
    
    for x,y in zip(X_binarised_train, Y_train):
        y_pred = np.sum(x)>=b       #if summation of x(i) for all features is greater than threshold "b", then predict 1 else 0
        accurate_rows += (y==y_pred)              #Total number of accurate rows 
        accuracy=accurate_rows/X_binarised_train.shape[0]             
    print("  {:02d}    {}    {:.2f}".format(b,accurate_rows, accuracy))
    
    if(max_accurate_rows<accurate_rows):             #if max value less than current then update the values
        max_accurate_rows=accurate_rows
        b_optimum=b
    
        
print("b_optimum=",b_optimum)
print("max_accuracy=",max_accurate_rows/X_binarised_train.shape[0]*100)
```

## On test data

```python
test_accurate_rows = 0
test_max_accurate_rows=0
test_b_optimum=0

print("For Test data:\n")
print("OUTPUT FORMAT iS \n{b ,Accurate_Rows ,Accuracy }")
for b in range(1,31):
    y_pred=0
    test_accurate_rows=0
    
    for x,y in zip(X_binarised_test, Y_test):
        y_pred = np.sum(x)>=b
        test_accurate_rows += (y==y_pred)

    test_Accuracy=test_accurate_rows/X_binarised_test.shape[0]
    if(test_max_accurate_rows<test_accurate_rows):
        test_max_accurate_rows=test_accurate_rows
        test_b_optimum=b
    print("  {:02d}    {}    {:.2f}".format(b,test_accurate_rows, test_Accuracy))

print("test_b_optimum=",test_b_optimum)
print("test_max_accuracy=",test_max_accurate_rows/X_binarised_test.shape[0]*100)
```


# Vectorised Implementation
In the above i have implemented the model using For loops but as we know looping is costly in terms of computation so i will also implement it using vectors.
We can eliminate the inner for loop.

## Train data

```python
max_accurate_rows=0
print("For Train data:\n")
print("OUTPUT FORMAT iS \n{b ,Accurate_Rows ,Accuracy }")
for b in range(1,31):
    
    Sum= np.sum(X_binarised_train, axis=1)   #Summation of x(i) values for all features

    y_pred =  (Sum>= b)                      # prediction of y_pred values using threshold = b

    if y_pred.dtype==bool:             #Converting boolean values into integers
        y_pred= y_pred.astype(int)
    
    accurate_rows = np.count_nonzero(y_pred == Y_train)     #counting number of accurately predicted values
    accuracy = np.mean(y_pred == Y_train)                   #accuracy
    
    print("  {:02d}    {}    {:.2f}".format(b,accurate_rows, accuracy))
    
    if(max_accurate_rows<accurate_rows):
        max_accurate_rows=accurate_rows
        b_optimum=b
        
print("b_optimum=",b_optimum)
print("max_accuracy=",max_accurate_rows/X_binarised_train.shape[0]*100)
```

## Test Data

```python
max_accurate_rows=0
print("For Test data:\n")
print("OUTPUT FORMAT iS \n{b ,Accurate_Rows ,Accuracy }")
for b in range(1,31):
    
    Sum= np.sum(X_binarised_test, axis=1)   #Summation of x(i) values for all features

    y_pred =  (Sum>= b)                      # prediction of y_pred values using threshold = b

    if y_pred.dtype==bool:             #Converting boolean values into integers
        y_pred= y_pred.astype(int)
    
    accurate_rows = np.count_nonzero(y_pred == Y_test)
    accuracy = np.mean(y_pred == Y_test)
    
    print("  {:02d}    {}    {:.2f}".format(b,accurate_rows, accuracy))
    
    if(max_accurate_rows<accurate_rows):
        max_accurate_rows=accurate_rows
        b_optimum=b
        
print("b_optimum=",b_optimum)
print("max_accuracy=",max_accurate_rows/X_binarised_test.shape[0]*100)
```

```python
from sklearn import metrics
Sum= np.sum(X_binarised_test, axis=1)   #Summation of x(i) values for all features
y_pred =  (Sum>= 28)
cm = metrics.confusion_matrix(Y_test, y_pred)  #false-false, true-false, false-true and true-true values 
print(cm)
```

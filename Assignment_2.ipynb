'O' refers to Object data type.

## Task-6 : Display the ‘Employment’ of customers who mostly ‘Rent’

```python
#Firstly we locate all the data rows having householder variable value to be 'RENT' and save the new dataframe in rent

rent = data.loc[data['householder']=='RENT']

rent['employment'].value_counts()    #we count the occurance of every unique value of employment
```

From the output we can conclude that Teachers and Civil servants are the maximum who rent.

But as we are asked to give the employment type of those who MOSTLY rent, we can select say top 3 as our employment type.

```python
#We will display only top 3 employment types who mostly rent
print("The Employment of customers who mostly rent are {}, {} and {}".format
    (rent['employment'].value_counts().idxmax(), #to retrieve max value
    rent['employment'].value_counts().index[1],      #retrieve index no.1
    rent['employment'].value_counts().index[2]))     #retrieve index no.2
```
